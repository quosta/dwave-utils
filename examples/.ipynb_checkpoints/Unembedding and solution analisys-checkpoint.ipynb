# Unembedding and solution analysis for small QUBOs

## Let's start with a dense BQM embedded on the 2000Q QPU

```python
import dwave_utils.utils as dwu 

bqm = dwu.random_dense_bqm(7)

new_bqm, emb = dwu.embedded_bqm(bqm, 
                                target='advantage',      # Advantage QPU
                                chain_strength=None,     # use the default uniform_torque_compensation
                                clique = False,          # do not use minorminer clique embedding
                                return_embedding=True) 
```

## The BQM and the embedded BQM can be solved via simulated annealing:

```python
from neal import SimulatedAnnealingSampler

sampler = SimulatedAnnealingSampler()

bqm_sol = sampler.sample(bqm,num_reads=100)
print('bqm: ', bqm_sol.first)

new_bqm_sol = sampler.sample(new_bqm,num_reads=100)
print('new_bqm: ', new_bqm_sol.first)
```

## If the BQM is small enough we can "bruteforce" the solution, by trying all the possible bitstrings. I will show both the old and the new BQM results:

```python
sol_df = dwu.bruteforce_bqm(bqm)
new_sol_df = dwu.bruteforce_bqm(new_bqm)

print(sol_df,new_sol_df)
```

## The energy found is the same, but what is interesting is that we can plot the entire energy landscape:

```python
import seaborn as sns
import matplotlib.pyplot as plt

x = sol_df.index
y = sol_df.energy
x_new = new_sol_df.index
y_new = new_sol_df.energy
size = 15

fig = plt.figure(figsize=(size, size))
gs = fig.add_gridspec(2, 1)

with sns.axes_style("whitegrid"):
    ax = fig.add_subplot(gs[0, 0])
    sns.scatterplot(x=x,y=y)
    sns.lineplot(x=x,y=y,linestyle='dotted')
    
with sns.axes_style("whitegrid"):
    ax = fig.add_subplot(gs[1, 0])
    sns.scatterplot(x=x_new,y=y_new)
    sns.lineplot(x=x_new,y=y_new,linestyle='dotted')
```

## I can now sort the solution from lowest energy to highest energy:

```python
plt.figure(figsize=(size,size))
new_sol_df = new_sol_df.sort_values('energy')
new_sol_df = new_sol_df.reset_index()
x_new = new_sol_df.index
y_new = new_sol_df.energy

sol_df = sol_df.sort_values('energy')
sol_df = sol_df.reset_index()
x = sol_df.index
y = sol_df.energy
    
fig = plt.figure(figsize=(size, size//2))
gs = fig.add_gridspec(1,2)

with sns.axes_style("whitegrid"):
    ax = fig.add_subplot(gs[0, 0])
    sns.scatterplot(x=x,y=y,marker='x')
    sns.lineplot(x=x,y=y,linestyle='dotted')
    
with sns.axes_style("whitegrid"):
    ax = fig.add_subplot(gs[0, 1])
    sns.scatterplot(x=x_new,y=y_new,marker='x')
    sns.lineplot(x=x_new,y=y_new,linestyle='dotted')
```

## We can calculate the Hamming distance between the optimal solution and all the other solutions, sorted in increasing energy. Notice how the distance increases as energy quality is depleted:

```python
from scipy.spatial.distance import hamming

idx_plus = sol_df[1:].iterrows()
ham = [0]
for i, row in idx_plus:
    bitstr = row[1]
    result = int(hamming(sol_df.loc[0].bitstr,bitstr) * len(bitstr))
    ham.append(result)    
sol_df['hamming'] = ham

idx_plus = new_sol_df[1:].iterrows()
ham = [0]
for i, row in idx_plus:
    bitstr = row[1]
    result = int(hamming(new_sol_df.loc[0].bitstr,bitstr) * len(bitstr))
    ham.append(result)    
new_sol_df['hamming'] = ham



x = sol_df.index
y = sol_df.hamming
x_new = new_sol_df.index
y_new = new_sol_df.hamming

fig = plt.figure(figsize=(size, size//2))
gs = fig.add_gridspec(1,2)

with sns.axes_style("whitegrid"):
    ax = fig.add_subplot(gs[0, 0])
    sns.scatterplot(x=x,y=y,marker='x')
    sns.lineplot(x=x,y=y,linestyle='dotted')
    
with sns.axes_style("whitegrid"):
    ax = fig.add_subplot(gs[0, 1])
    sns.scatterplot(x=x_new,y=y_new,marker='x')
    sns.lineplot(x=x_new,y=y_new,linestyle='dotted')
```

## I want to analyze a model where the embedding was bad. For example let's embed a BQM with a very light chain strength and sample the solution with SA:

```python
bqm = dwu.random_dense_bqm(8)

new_bqm, emb = dwu.embedded_bqm(bqm, 
                                target='2000',          
                                chain_strength=0.1,     # bad chain strength
                                return_embedding=True) 

bqm_sol = sampler.sample(bqm,num_reads=1000)
print('bqm: ', bqm_sol.first)

new_bqm_sol = sampler.sample(new_bqm,num_reads=1000)
print('new_bqm: ', new_bqm_sol.first)
```

## The new_bqm has a energy even lower that the original! This is possible only because a chain break has occurred. I will try to resolve the chain break by producing a new state that has the same value as most of the qubits of the broken chain:

```python
fixed_sol = dwu.majority_voting(emb,
                                new_bqm_sol.first.sample)
```

```python
print('Unembedded BQM: ', dwu.get_energy_from_solution(bqm,bqm_sol.first.sample))
print('Embedded BQM with broken chains: ', dwu.get_energy_from_solution(new_bqm,new_bqm_sol.first.sample))
print('Embedded BQM with fixed chains: ', dwu.get_energy_from_solution(new_bqm,fixed_sol))
```

## The fixed solution is not as low as the unembedded one, but now that the chains are coherent we can unembed the solution and see how many qubits we got wrong:

```python
print("Fixed solution mapped back to the unembedded BQM: \n",dwu.get_unembedded_solution(emb,fixed_sol).values())
print("Best solution from the unembedded BQM: \n",bqm_sol.first.sample.values())
```
